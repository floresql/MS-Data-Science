### Cleaning Data

#### Account for missing values

```python
df.replace([None], np.nan, inplace=True)
```

```python
clean_df = df.copy()
```

```python
clean_df.CustomerNo.unique()
```

```python
clean_df[clean_df.CustomerNo.isna()]
```

```python
df_missing = clean_df[clean_df.CustomerNo.isna()].copy()
```

```python
df_missing['CustomerNo'].fillna(0, inplace=True)
```

Selecting by index lable vs selecting by row

```python
# making a copy of our df
year_df = df.copy()

# creating an index lable
year_df.set_index('year', inplace=True)
```

```python
year_df.loc[1952].head(10) 
```

Filtering and aggragation

```python
df.groupby('year')['lifeExp'].mean()
```

Breaking up long lines with \ at the end

```python
df.groupby(['year', 'continent']) \
[['lifeExp', 'gdpPercap']]. \
mean().head(n=20)
```

Flatten results

```python
df.groupby(['year', 'continent']) \
[['lifeExp', 'gdpPercap']]. \
mean(). \
reset_index().\
head(15)
```

Count unique

```python
df.groupby('continent')['country'].nunique()
```

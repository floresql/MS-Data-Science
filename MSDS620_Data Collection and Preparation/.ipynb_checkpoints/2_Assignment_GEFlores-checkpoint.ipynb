## Task 1:

A) Create 3 related dataframes with "several" values each:
   - employees - contains a list of employees, a department id, and branch id for each.

```python
import pandas as pd
```

```python
# Create the employees df
data_employees = {
    'employee_id': [1, 2, 3, 4, 5],
    'employee_name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'department_id': [101, 102, 101, 103, 102],
    'branch_id': [201, 202, 201, 201, 202]
}
employees_df = pd.DataFrame(data_employees)

# Display df
employees_df
```

   - departments - list of department names and ids.

```python
# Create the departments df
data_departments = {
    'department_id': [101, 102, 103],
    'department_name': ['Engineering', 'Sales', 'Marketing']
}
departments_df = pd.DataFrame(data_departments)

# Display df
departments_df
```

   - branch office - list of branch office names and ids.

```python
# Create the branch office df
data_branches = {
    'branch_id': [201, 202],
    'branch_name': ['Rochester', 'Tampa']
}
branch_office_df = pd.DataFrame(data_branches)


# Display df
branch_office_df
```

B) Demonstrate several of the dataframe merges (see https://chrisalbon.com/python/data_wrangling/pandas_join_merge_dataframe/ for help):
   - left join (2 tables)

```python
# left join
left_join_df = pd.merge(employees_df, departments_df, on='department_id', how='left')

left_join_df

```

   - right join (2 tables)

```python
# Right Join
right_join_df = pd.merge(employees_df, departments_df, on='department_id', how='right')

right_join_df
```

   - outer join (2 tables)

```python
# Outer Join
outer_join_df = pd.merge(employees_df, departments_df, on='department_id', how='outer')

outer_join_df
```

   - inner join (2 tables)

```python
# Inner Join
inner_join_df = pd.merge(employees_df, departments_df, on='department_id', how='inner')

inner_join_df
```

   - Perform a join that correctly shows a list of employees, their department, and their branch.
     * There are multiple ways this could be done, including doing it in 2 steps, chaining joins, simultaneous left and right join, etc.

```python
# List of employees, their department, and their branch

# First merge employees with departments
employee_dept_df = pd.merge(employees_df, departments_df, on='department_id', how='inner')

# Then merge the result with branch offices
employee_dept_branch_df = pd.merge(employee_dept_df, branch_office_df, on='branch_id', how='inner')

employee_dept_branch_df
```

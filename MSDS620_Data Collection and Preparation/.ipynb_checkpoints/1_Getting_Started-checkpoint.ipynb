## Selecting rows and columns

### Single Column

Selecting subsets of your data is a common task. First, let's isolate a single column:

````python
country_df = df['country']
country_df.head(10)
````

````python
country_df.tail(10)
````

Accessing a single column works much like a Python dictionary. In fact, we will see in Chapter 2 that pandas dataframes and dictionaries are related.

### Multiple Columns

Now, let's look at a subset of columns. The syntax is a little weird, until you realize that the outer brackets tell the dataframe about a column, and the inner set denotes a list of columns.

````python
subset = df[['country', 'continent', 'year']]
subset.head()
````

````python
subset.shape
````

We could just as easily have passed in a list:

````python
sub_columns = ['country', 'continent', 'year']
sub_df = df[sub_columns]
sub_df.head()
````

### Rows - loc[] and iloc[]

Initially, loc[] and iloc[] can be confusing. The important factor is to remember that while the leftmost column *looks* like an index number, **it isn't**. It is a number by default *unless* you have specified otherwise.

Type | Description
-----|------------
loc[] | select by index *label* (number by default)
iloc[] | select by row number (true index number, regardless of label)

````python
df.head()
````

````python
df.loc[0]
````

````python
df.iloc[0]
````

Looking at the two cells above, it is hard to tell the difference. Let's copy df and give it a new index to help illustrate.

````python
year_df = df.copy()
````

````python
year_df.set_index('year', inplace=True)
````

````python
year_df.head()
````

Notice how the numbers on the left are gone, replaced by the year column. Now, we will compare loc[] and iloc[].

````python
year_df.loc[1952].head(10) # head() is chained to keep the output to a manageable number.
````

````python
year_df.iloc[0]
````

Notice how we only returned one row but the data was given as a table with column names on the left and data on the right. 

### Multiple rows

We already saw one example of multiple rows above, where our loc[] label matched multiple rows. Here are a couple more ways to get many rows.

````python
year_df.iloc[[0,10,100]]
````

````python
index_range = list(range(0,10))
````

````python
year_df.iloc[index_range]
````

### Column slicing

A "slice" of a data structure is some contiguous subset of that structure. For example, a simple list:

````python
index_range
````

Slice syntax can be considered as \["start element" : "number of elements"\]. So, if you want 5 element, starting at 0:

````python
index_range[0:5] # 0 to 4 is five elements
````

````python
index_range[:] # all elements
````

Two dimensional objects like like a matrix (M) are a little different. The general syntax is<br>
`M['row start' : 'row end', 'col start' : 'col end']`

type | syntax
-----| ------
row slice | M\[start:stop, : \]
column slice | M\[:, start:stop \]

````python
# Labels 0 through 4, all columns
df.loc[0:4, :] # Could also use iloc
````

````python
# All rows, columns year and pop
df.loc[:, ['year', 'pop']].head(10)
````

````python
# All rows, first 3 columns
df.iloc[:, :3].head() # 0 is assumed as starting column number
````

## Filtering and aggragation

Dataframes have a 'groupby()' function. Can you guess what it does?

Let's find out how average life expectancy changes over the years:

````python
df.groupby('year')['lifeExp'].mean()
````

Long lines can be broken up by putting a \ character at the end:

````python
df.groupby(['year', 'continent']) \
[['lifeExp', 'gdpPercap']]. \
mean().head(n=20)
````

#### Flatten out the results

````python
df.groupby(['year', 'continent']) \
[['lifeExp', 'gdpPercap']]. \
mean(). \
reset_index().\
head(15)
````

#### Count unique

````python
df.groupby('continent')['country'].nunique()
````

## Python data structures

At the risk of boring you, let's go back to basics. A **variable** is a memory location to store a discrete piece of information: a name, an date, the number of widgets produced in 2014, etc. But what happens if you have **five names**? Do you make 5 name variables?

```
name1 = 'Bob'
name2 = 'Cheryl'
name3 = 'Dave'
name4 = 'Jenny'
name5 = 'Waldo'
```
OK, fine. What if you have 1,000 names? 10,000? What if you don't know how many names you will have? In cases where we have 2 or more similar items, we generally want to put them into a container for easier handling. Think about it -- is it easier to carry around a spare left shoe and right shoe in your hands or put them in a box and carry the box?

In Python, there are two major types of built-in container: the **list** and the **dictionary**. 
* List - A contiguous group of memory spaces accessed by an index number. If the container below is a list called **names** then names\[1\] holds "Cheryl."

<img style="padding-right:10px;" src="figures_wk1/pill_box_list.png" width="450"><br>

* Dictionary - Similar to a list, except each *value* is accessed by a *key*. If the container below is a dictionary called **staff** then staff\["HR"\] would answer the age-old question, "Where is Waldo?"

<img style="padding-right:10px;" src="figures_wk1/pill_box_dictionary.png" width="450"><br>

````python
names = ["Bob","Cheryl","Dave","Jenny", "Waldo"]
names[1]
````

````python
staff = {'QA':"Bob",'Manager':"Cheryl","Dev":"Dave","Marketing":"Jenny", 'HR':"Waldo"}
staff['HR']
````

## Pandas series and dataframe

The Pandas series is roughly analogous to a list and dataframes are like dictionaries.

````python
name_series = pd.Series(["Bob","Cheryl","Dave","Jenny", "Waldo"])
name_series
````

````python
name_series[1]
````

````python
type(name_series)
````

````python
staff_df = pd.DataFrame({'QA':["Bob"],'Manager':["Cheryl"],"Dev":["Dave"],"Marketing":["Jenny"], 'HR':["Waldo"]})
staff_df
````

Notice in the dataframe above that the dictionary keys became the column names and the values needed to be lists to account for dataframe rows. We could just as easily have passed multiple values to make more rows, like so:

````python
staff_df = pd.DataFrame({'QA':["Bob","Gina"],'Manager':["Cheryl","Tom"],"Dev":["Dave","Susan"],"Marketing":["Jenny","Richard"], 'HR':["Waldo","Mario"]})
staff_df
````

## Boolean Operations

### Series

We've seen how to subset by rows and columns. Now we are going to see how to add boolean filtering logic to those operations. 

First, I'll load the same data file the book uses. Here I'm getting it from the data subdirectory of the author's GitHub repository I cloned.

````python
scientists = pd.read_csv("data_wk1/scientists.csv")
scientists.head()
````

Following the book, let's make a Series out of the Age column:

````python
ages = scientists['Age']
ages.head()
````

````python
type(ages)
````

````python
ages.mean()
````

What happens when we apply a boolean operation to our series?

````python
ages > ages.mean()
````

So, a boolean operation gives boolean output. Now, what happens if we use that when indexing?

````python
ages[ages > ages.mean()]
````

Hmmmm... That's interesting. Let's convert that bunch of True/False into a list and see if we can use that in the brackets. Wrapping our boolean operation in `list()` and storing it in a variable should do the trick:

````python
# Capture output of boolean as input to a list constructor, store the result in a variable
oldies = list(ages > ages.mean()) 

# Check the contents of the variable to make sure we got what we think got
oldies
````

````python
ages[oldies]
````

Because of underlying efficiencies in the way data is stored and handled, we can perform operations on entire series at once.

````python
ages + 10
````

````python
ages * 2
````

````python
ages + 'a'
````

<div class="alert alert-block alert-info">
<b>What happened?</b> We can't add a string ('a') to ages, becaues ages is numeric.
</div>

## Changing data 

### Adding a column

The born and died columns in our dataframe are lying to you. They *look* like a date and time but they are not.

````python
scientists.info()
````

They are type *object*, meaning they are strings, just like the name column. Many times this won't matter so much... until you want to do some time series analysis. Then it is vital to get the type right.

To correct this problem, we can use the Pandas to_datetime() function to do the conversion and we will store the new columns in variables.

````python
born_date = pd.to_datetime(scientists['Born'], format='%Y-%m-%d')
born_date
````

That formatting with the % hearkens back to C programming days. We will see more examples of that in Chapter 11. 

````python
died_date = pd.to_datetime(scientists['Died'], format='%Y-%m-%d')
type(died_date)
````

To add those new series as columns, all we have to do is use a new name in the Dataframe brackets and set it equal to the series. The book demonstrates "multiple assignment" with this statement:

`scientists['born_dt'], scientists['died_dt'] = born_date, died_date`

But to demonstrate that you aren't **_required_** to use multiple assignment, I'll do it on two lines below:

````python
scientists['born_dt'] = born_date
scientists['died_dt'] = died_date
scientists
````

````python
scientists.info()
````

### Changing in place

Generally, when we make changes to a Series or Dataframe, there are two ways to go about it - temporarily (meaning the operation 'returns' the change but the underlying data is not changed), or 'in place.' 

The book uses shuffling of the rows as an example however that doesn't seem useful in too many situations. Let's look at sorting instead. We will sort the Dataframe by when they died.

First, returning a temporary change:

````python
scientists.sort_values(by='died_dt')
````

We see the output of that command so it is something we could store in a variable. The underlying data is unchanged:

````python
scientists
````

Now, let's actually change the Dataframe:

````python
scientists.sort_values(by='died_dt', inplace=True)
````

````python
scientists
````

### Dropping columns and values

Pandas has a variety of commands to eliminate rows with nulls or entire columns. I'll copy the Age column specifically so I can drop it:

````python
scientists['Age_copy'] = scientists['Age'].copy()
scientists
````

````python
scientists.drop(['Age_copy'], inplace=True, axis=1)  # Axis=1 means column
scientists
````

Let's set all the ages above the mean to be null, and then drop the null rows.

````python
import numpy as np
scientists.loc[scientists['Age'] > scientists['Age'].mean(), 'Age'] = np.nan
scientists
````

````python
scientists.dropna() # Notice we did not set inplace=True so this is temporary
````

Let's put the ages back. This is tricky -- datetimes are tough to work with.

In the line below, a "timedelta64" is the result of subtracting two datetimes and the \[Y\] means I want only the year part. 

````python
scientists
````

````python
scientists['Age'].fillna(value=(scientists['died_dt'].dt.year - scientists['born_dt'].dt.year), inplace=True)
scientists
````

````python
scientists.info()
````

## Saving and retrieving

Since the book does a good job of explaining pickle, CSV, etc., we will briefly look at two common formats for storing large, tabular data: HDF5 and Feather.

### HDF5

One problem with storing complex data is that you have to know the structure of the data to read it back in. According to the Pandas documentation, HDF stands for "Hierarchical Data Format." HDF is self-describing, meaning applications do not have to know the data structure to work with the files. Also, one HDF file can hold multiple objects (dataframes or series). HDF is capable of storing multi-gigabyte and larger files, with or without compression, and extremely fast read and write times.

We will write our scientists out to a file, then read it back into a different dataframe to make sure it loaded properly.

````python
# HDF5 Requires each object stored must have a "key" to save and read back.
scientists.to_hdf("scientists.h5", key='scientists', mode='w')
````

````python
new_sci = pd.read_hdf('scientists.h5', key='scientists')
new_sci
````

### Feather

Feather is a relatively new file format created by the author of Pandas and utilizing Apache Arrow (https://arrow.apache.org/) for storing columnar data. Feather is designed to be interoperable with both Python and R and has extremely fast load and save times.

The Pandas implementation of to_feather() can't deal with the mixed-up index, so let's reset it.

````python
scientists[scientists['index']==0]
````
